```python
# Load CSV 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("country_deatils_Director.csv")
```

```python
print(df.head(10))
print(df.isna().sum())
```

```python
country_df = df.copy()
country_df.columns = [c.strip().lower() for c in country_df.columns]
```

```python
# Trim string fields
for col in ['name', 'department']:
    country_df[col] = country_df[col].astype(str).str.strip()
```

```python
# Coerce numeric fields safely
for col in ['id', 'gender', 'uid']:
    country_df[col] = pd.to_numeric(country_df[col], errors='coerce')
```

```python
# Basic validity rules
country_df.loc[country_df['gender'].isin([0, 1, 2]) == False, 'gender'] = np.nan
```

```python
# Remove duplicates if any
country_df = country_df.drop_duplicates()
```

```python
print(country_df.head(10))
print(country_df.isna().sum())
print(country_df.shape)
```

```python
# Quick exploration plots: gender distribution and top names (as a sanity check)
import matplotlib
import matplotlib.pyplot as plt
import warnings

# stop font warnings
warnings.filterwarnings("ignore", category=UserWarning, module="matplotlib")

# force default font
matplotlib.rcParams['font.family'] = 'DejaVu Sans'
matplotlib.rcParams['font.sans-serif'] = ['DejaVu Sans']

```

```python
plt.style.use('default')
plt.rcParams['font.family'] = ['Inter']
plt.rcParams['font.sans-serif'] = ['Inter']
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=country_df, x='gender', color='#2563EB')

plt.title("Gender Distribution", fontsize=16, weight='bold')
plt.grid(axis='y', alpha=0.3)

sns.despine()
plt.show()
```

```python
# Top repeated names (sanity check)
name_counts = country_df['name'].value_counts().head(15)

plt.figure(figsize=(10,6))
sns.barplot(x=name_counts.values, y=name_counts.index, color='#2563EB')

plt.title("Top Repeated Names (Sanity Check)", fontsize=14, weight='bold')
plt.xlabel("Count")
plt.ylabel("Name")
plt.grid(axis='x', alpha=0.3)
sns.despine()
plt.show()

# quick checks
print(country_df['gender'].value_counts(dropna=False))
print(country_df['department'].value_counts(dropna=False).head())
```
